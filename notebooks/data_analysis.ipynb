```python
folder_path = "../data/20181029_D10_0900_0930/Annotations"

# Get all files in the folder
all_files = os.listdir(folder_path)

# Filter for files with a ".csv" extension
csv_files = [file for file in all_files if file.endswith('.csv')]

print(len(csv_files))
```

```python
common_path = "../data/20181029_D10_0900_0930/Annotations"
dfs = []

# Read each CSV file and append its DataFrame to the list
for file_name in csv_files:
    df = pd.read_csv(f'{common_path}/{file_name}', sep="[,;:]", index_col=False)
    dfs.append(df)

# Combine all DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# Print or use the combined DataFrame as needed
print(combined_df)
```
